**[Task 2.1.]()** Load and show the image:

```python
lenarray = plt.imread('data/lena.png')
plt.imshow(lenarray, cmap='gray')
```

**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
sobel_v = np.array([
    [-1,  0,  1],
    [-2,  0,  2],
    [-1,  0,  1]
])

sobel_h = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
])

output_v = np.zeros_like(lenarray)
output_h = np.zeros_like(lenarray)
```

**[Task 2.3.]()** Test your implementations from above:

```python

for x in range(2, len(lenarray)-2) :
    for y in range(2, len(lenarray)-2) :
        local = lenarray[x-1:x+2, y-1:y+2]
        output_v[x, y] = np.sum(local * sobel_v)
        

for x in range(2, len(lenarray)-2) :
    for y in range(2, len(lenarray)-2) :
        local = lenarray[x-1:x+2, y-1:y+2]
        output_h[x, y] = np.sum(local * sobel_h)

plt.subplot(1,2,1)        
plt.imshow(output_v, cmap='gray')


plt.subplot(1,2,2)    
plt.imshow(output_h, cmap='gray')
```

**[Task 2.4.]()** Compute the image gradient magnitude:

```python
# Magnitude

magn = np.sqrt(output_v ** 2 + output_h ** 2)
print(magn)
plt.imshow(magn, cmap = 'gray')
magn_mächt = np.sum(magn)
print(magn_mächt)
```
